```python
import tkinter as tk
from tkinter import *


janela = tk.Tk()
janela.title("Calculadora de media")

nota1 = None
nota2 = None
nota3 = None
nota4 = None
notas = []


while True:

    #inseri um campo de entrada na janela tk
    nota1 = tk.Entry(janela, width=3)
    nota1.grid(column=0, row=1, padx=5, pady=5)
    notas.append(nota1)

    # Verificar se o usuário deixou em branco
    if nota1.get() == "":
        break

    try:
        # Tentar converter a entrada do usuário em um número inteiro
        nota1 = int(nota1.get())
        notas.append(nota1)

        # Se a conversão for bem-sucedida, interromper o loop e continuar com o código
        break
    except ValueError:
        # Se a conversão falhar, imprimir uma mensagem de erro e continuar o loop
        print("Por favor, digite um número inteiro válido ou deixe em branco.")

while True:
    # Pedir ao usuário para digitar a segunda nota, permitindo deixar em branco
    nota2 = tk.Entry(janela, width=3)
    nota2.grid(column=1, row=1, padx=5, pady=5)
    notas.append(nota2)

    # Verificar se o usuário deixou em branco
    if nota2.get() == "":
        break

    try:
        # Tentar converter a entrada do usuário em um número inteiro
        nota2 = int(nota2.get())
        notas.append(nota2)

        # Se a conversão for bem-sucedida, interromper o loop e continuar com o código
        break
    except ValueError:
        # Se a conversão falhar, imprimir uma mensagem de erro e continuar o loop
        print("Por favor, digite um número inteiro válido ou deixe em branco.")


while True:
    # Pedir ao usuário para digitar a terceira nota, permitindo deixar em branco
    nota3 = tk.Entry(janela, width=3)
    nota3.grid(column=2, row=1, padx=5, pady=5)
    notas.append(nota3)

    # Verificar se o usuário deixou em branco
    if nota3.get() == "":
        break

    try:
        # Tentar converter a entrada do usuário em um número inteiro
        nota3 = int(nota3.get())
        notas.append(nota3)

        # Se a conversão for bem-sucedida, interromper o loop e continuar com o código
        break
    except ValueError:
        # Se a conversão falhar, imprimir uma mensagem de erro e continuar o loop
        print("Por favor, digite um número inteiro válido ou deixe em branco.")


while True:
    # Pedir ao usuário para digitar a quarta nota, permitindo deixar em branco
    nota4 = tk.Entry(janela, width=3)
    nota4.grid(column=3, row=1, padx=5, pady=5)
    notas.append(nota4)
    # Verificar se o usuário deixou em branco
    if nota4.get() == "":

        break

    try:
        # Tentar converter a entrada do usuário em um número inteiro
        nota4 = int(nota4.get())
        notas.append(nota4)

        # Se a conversão for bem-sucedida, interromper o loop e continuar com o código
        break
    except ValueError:
        # Se a conversão falhar, imprimir uma mensagem de erro e continuar o loop
        print("Por favor, digite um número inteiro válido ou deixe em branco.")

resultado_label = tk.Label(janela, text="")
resultado_label.grid(column=1, row=5, columnspan=2, padx=10, pady=10)


def calcular_media(notas):
    valores = []
    for nota in notas:
        valor = nota.get()
        if valor == "":
            continue
        partes = valor.split(",")
        if len(partes) == 2:
            inteiro, decimal = partes
            valor = "{}.{}".format(inteiro, decimal)
        valores.append(float(valor))
    if valores:
        media = sum(valores) / len(valores)
        return media
    else:
        resultado_label.config(text="você precisa digitar suas notas!")
        return ""


#cria
nota1_text = Label(janela, text="Nota 1")
nota1_text.grid(row=0, column=0, padx=10, pady=10)

nota2_text = Label(janela, text="Nota 2")
nota2_text.grid(row=0, column=1, padx=10, pady=10)

nota3_text = Label(janela, text="Nota 3")
nota3_text.grid(row=0, column=2, padx=10, pady=10)

nota4_text = Label(janela, text="Nota 4")
nota4_text.grid(row=0, column=3, padx=10, pady=10)


def resultado():
    calc = calcular_media(notas)

    if calc == "" or 0:
        resultado_label.config(text="você precisa digitar suas notas!")
    else:
        print("calculou")
        resultado_label.config(text="A média é: {:.2f}".format(calc))


def atualiza_botao():
    if len(notas) == "":
        button.config(command=lambda: None)
        print("a lista notas está vazia.")
    else:

        button.config(command=resultado)
        print("botão atualizado")


button = tk.Button(janela, text="Calcular", width=10, height=1,
                   command=lambda: (atualiza_botao(), resultado()))
button.grid(column=1, row=2, columnspan=2, padx=10, pady=25)


#termina o codigo
janela.mainloop()
```
